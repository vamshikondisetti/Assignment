# _Hierarchical

```python
# import hierarchical clustering libraries
import scipy.cluster.hierarchy as sch  # create dendrogram
from sklearn.cluster import AgglomerativeClustering

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
crime = pd.read_csv("crime_data.csv")
```

```python
crime
```

```python
# Normalization function # use sklearn library norm
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return x
```

```python
# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(crime.iloc[:,1:])
```

```python
df_norm
```

```python
# create dendrogram
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='complete'))
```

```python
# create clusters
CR_D = AgglomerativeClustering(n_clusters=5, affinity = 'euclidean', linkage = 'complete')
```

```python
CR_D
```

```python
CR_D.fit(df_norm)
# hc.predict(df_norm)
```

```python
# save clusters for chart
y_CR_D = CR_D.fit_predict(df_norm)
# Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
y_CR_D
```

```python
crime['h_clusterid'] = y_CR_D
```

```python
crime.head()
```

```python
crime1=crime.sort_values("h_clusterid")
crime1.iloc[:,[0,-1]]
```

# K_Means

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
crimedata = pd.read_csv("crime_data.csv")
```

```python
crimedata
```

```python
# Standardisation function
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_crime_df = scaler.fit_transform(crimedata.iloc[:,1:])
```

```python
scaled_crime_df
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_crime_df)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Build Cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(4, random_state=42)
clusters_new.fit(scaled_crime_df)
```

```python
clusters_new.labels_
```

```python
#Assign clusters to the data set
crimedata['clusterid_new'] = clusters_new.labels_
```

```python
crimedata
```

```python
crimedata.groupby('clusterid_new').agg(['mean']).reset_index()
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
predict=kmeans.fit_predict(scaled_crime_df)
```

```python
predict
```

```python
silhouette_score_average = silhouette_score(scaled_crime_df, predict)
```

```python
silhouette_score_average
```

```python
silhouette_score_average1 = silhouette_score(scaled_crime_df, clusters_new.predict(scaled_crime_df))
```

# DBSCAN

```python
#Import the libraries
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("crime_data.csv")
```

```python
df
```

```python
print(df.info())
```

```python
df.drop(['Name'],axis=1,inplace=True)
df
```

```python
stscaler = StandardScaler()
X = stscaler.fit_transform(df)
```

```python
X
```

```python
dbscan = DBSCAN(eps=2, min_samples=2)
dbscan.fit_predict(X)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
cl
```

```python
clustered = pd.concat([df,cl],axis=1)
```

```python
clustered
```

```python
clustered[clustered['cluster']==-1]
```
